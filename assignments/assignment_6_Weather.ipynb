```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

import csv

```python
file_path = "knock.csv"
df = pd.read_csv(file_path, skiprows=22, delimiter=',', on_bad_lines='skip')
df.head()

```

```python
print(df.columns)
print(df.dtypes)
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
import matplotlib.dates as mdates

#Plot 1: The temperature over time
plt.figure(figsize=(12, 6), facecolor='lavender')
plt.plot(df.index, df['temp'], color='purple', linewidth=2, label='Temperature')
plt.xlabel('Time', fontsize=14, fontweight='bold')
plt.ylabel('Temperature (Â°C)', fontsize=14, fontweight='bold')
plt.title('Hourly Temperature Over Time', fontsize=16, fontweight='bold', pad=20)
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.xticks(rotation=25, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, which='both', axis='y', linestyle='--', linewidth=1.2, alpha=0.8)
plt.gca().set_facecolor('white')
for spine in ['top', 'bottom', 'left', 'right']:
    plt.gca().spines[spine].set_color('lightgreen')
    plt.gca().spines[spine].set_linewidth(2)
plt.legend(loc='upper left', fontsize=10)
plt.show()

#Plot 2: The mean temperature each day
daily_mean_temp = df['temp'].resample('D').mean()
plt.figure(figsize=(12, 6), facecolor='lavender')
plt.plot(daily_mean_temp.index, daily_mean_temp, color='orange', linewidth=2, label='Daily Mean Temp')
plt.xlabel('Time', fontsize=14, fontweight='bold')
plt.ylabel('Temperature (Â°C)', fontsize=14, fontweight='bold')
plt.title('Daily Mean Temperature Over Time', fontsize=16, fontweight='bold', pad=20)
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.xticks(rotation=25, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, which='both', axis='y', linestyle='--', linewidth=1.2, alpha=0.8)
plt.gca().set_facecolor('white')
for spine in ['top', 'bottom', 'left', 'right']:
    plt.gca().spines[spine].set_color('lightgreen')
    plt.gca().spines[spine].set_linewidth(2)
plt.legend(loc='upper left', fontsize=10)
plt.show()

#Plot 3: The mean temperature for each month
monthly_mean_temp = df['temp'].resample('M').mean()
plt.figure(figsize=(12, 6), facecolor='lavender')
plt.plot(monthly_mean_temp.index, monthly_mean_temp, color='blue', linewidth=2, label='Monthly Mean Temp')
plt.xlabel('Time', fontsize=14, fontweight='bold')
plt.ylabel('Temperature (Â°C)', fontsize=14, fontweight='bold')
plt.title('Monthly Mean Temperature Over Time', fontsize=16, fontweight='bold', pad=20)
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.xticks(rotation=25, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, which='both', axis='y', linestyle='--', linewidth=1.2, alpha=0.8)
plt.gca().set_facecolor('white')
for spine in ['top', 'bottom', 'left', 'right']:
    plt.gca().spines[spine].set_color('lightgreen')
    plt.gca().spines[spine].set_linewidth(2)
plt.legend(loc='upper left', fontsize=10)
plt.show()
```

```python
# Plot 1: The Windspeed (raw data)
plt.figure(figsize=(12, 6), facecolor='lavender')
plt.plot(df.index, df['wdsp'], color='blue', linewidth=1, label='Windspeed')
plt.xlabel('Time', fontsize=14, fontweight='bold')
plt.ylabel('Windspeed (knots)', fontsize=14, fontweight='bold')
plt.title('Hourly Windspeed Over Time', fontsize=16, fontweight='bold', pad=20)
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.xticks(rotation=25, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, which='both', axis='y', linestyle='--', linewidth=1.2, alpha=0.8)
plt.gca().set_facecolor('white')
for spine in ['top', 'bottom', 'left', 'right']:
    plt.gca().spines[spine].set_color('lightgreen')
    plt.gca().spines[spine].set_linewidth(2)
plt.legend(loc='upper left', fontsize=10)
plt.show()

# Plot 2: The rolling windspeed (24-hour window)
rolling_wdsp = df['wdsp'].rolling(window=24).mean()
plt.figure(figsize=(12, 6), facecolor='lavender')
plt.plot(df.index, rolling_wdsp, color='orange', linewidth=2, label='24-hour Rolling Windspeed')
plt.xlabel('Time', fontsize=14, fontweight='bold')
plt.ylabel('Windspeed (knots)', fontsize=14, fontweight='bold')
plt.title('24-hour Rolling Windspeed', fontsize=16, fontweight='bold', pad=20)
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.xticks(rotation=25, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, which='both', axis='y', linestyle='--', linewidth=1.2, alpha=0.8)
plt.gca().set_facecolor('white')
for spine in ['top', 'bottom', 'left', 'right']:
    plt.gca().spines[spine].set_color('lightgreen')
    plt.gca().spines[spine].set_linewidth(2)
plt.legend(loc='upper left', fontsize=10)
plt.show()

# Plot 3: The max windspeed for each day
daily_max_wdsp = df['wdsp'].resample('D').max()
plt.figure(figsize=(12, 6), facecolor='lavender')
plt.plot(daily_max_wdsp.index, daily_max_wdsp, color='green', linewidth=2, label='Daily Max Windspeed')
plt.xlabel('Time', fontsize=14, fontweight='bold')
plt.ylabel('Windspeed (knots)', fontsize=14, fontweight='bold')
plt.title('Daily Max Windspeed Over Time', fontsize=16, fontweight='bold', pad=20)
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.xticks(rotation=25, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, which='both', axis='y', linestyle='--', linewidth=1.2, alpha=0.8)
plt.gca().set_facecolor('white')
for spine in ['top', 'bottom', 'left', 'right']:
    plt.gca().spines[spine].set_color('lightgreen')
    plt.gca().spines[spine].set_linewidth(2)
plt.legend(loc='upper left', fontsize=10)
plt.show()

# Plot 4: The monthly mean of the daily max windspeeds
monthly_mean_daily_max_wdsp = daily_max_wdsp.resample('M').mean()
plt.figure(figsize=(12, 6), facecolor='lavender')
plt.plot(monthly_mean_daily_max_wdsp.index, monthly_mean_daily_max_wdsp, color='red', linewidth=2, label='Monthly Mean of Daily Max Windspeed')
plt.xlabel('Time', fontsize=14, fontweight='bold')
plt.ylabel('Windspeed (knots)', fontsize=14, fontweight='bold')
plt.title('Monthly Mean of Daily Max Windspeeds', fontsize=16, fontweight='bold', pad=20)
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.xticks(rotation=25, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, which='both', axis='y', linestyle='--', linewidth=1.2, alpha=0.8)
plt.gca().set_facecolor('white')
for spine in ['top', 'bottom', 'left', 'right']:
    plt.gca().spines[spine].set_color('lightgreen')
    plt.gca().spines[spine].set_linewidth(2)
plt.legend(loc='upper left', fontsize=10)
plt.show()
```
