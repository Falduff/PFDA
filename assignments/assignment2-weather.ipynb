```python
# Imports
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
import datetime as dt
```

Read in csv using pandas 

```python
csv = "weatherreadings1.csv"
df = pd.read_csv(csv)
```

Look at the dataframe structure

```python
df
```

Create a new dataframe called df_filtered that only contains the reportStartDateTime and dryBulbTemperature_Celsius columns.
Then change the datatype of the reportStartDateTime column to datetime without the timezone.

```python
df_filtered = df[['reportStartDateTime', 'dryBulbTemperature_Celsius']].copy()
df_filtered['reportStartDateTime'] = pd.to_datetime(df_filtered['reportStartDateTime']).dt.tz_localize(None)
```

Look at the structure of the new dataframe

```python
df_filtered
```

Check datatypes of the columns

```python
print(df_filtered['reportStartDateTime'].dtype)
print(df_filtered['dryBulbTemperature_Celsius'].dtype)
```

Check if there are any empty entries

```python
print(df_filtered.isnull().sum())
```

PLot the data from the filtered dataframe using matplotlib.
A scatter plot was used with time on our x-axis as the independent variable. And temperature on the y-axis as our dependent variable.
Dots were used to represent each reading and they were colour coded depending on the day the reading was taken.
A line was also added.
Some subtle colours were used to add some charm to the graph without being too distracting. 
Minor tweaks were made to legends, axes, borders, fonts etc to show the versitility of matplotlib as a plotting tool.

```python
# Create a figure for the plot with lavender outside the plot
plt.figure(figsize=(12, 6), facecolor='lavender')

# Plot the temperature line in purple
plt.plot(df_filtered['reportStartDateTime'], df_filtered['dryBulbTemperature_Celsius'], 
         color='purple', linewidth=2, marker='', label='Temperature Line')

# Separate data points for 02-10 and 03-10
# Filter the data for specific dates
df_02_10 = df_filtered[df_filtered['reportStartDateTime'].dt.strftime('%d-%m') == '02-10']
df_03_10 = df_filtered[df_filtered['reportStartDateTime'].dt.strftime('%d-%m') == '03-10']

# Plot the dots for 02-10 in red
plt.scatter(df_02_10['reportStartDateTime'], df_02_10['dryBulbTemperature_Celsius'], 
            color='red', label='02-10', s=15, zorder=5)  

# Plot the dots for 03-10 in blue 
plt.scatter(df_03_10['reportStartDateTime'], df_03_10['dryBulbTemperature_Celsius'], 
            color='blue', label='03-10', s=15, zorder=5)  

# Set labels and title with larger fonts and bold style
plt.xlabel('Time', fontsize=14, fontweight='bold')
plt.ylabel('Dry Bulb Temperature (Celsius)', fontsize=14, fontweight='bold')
plt.title('Temperature vs Time of Day', fontsize=16, fontweight='bold', pad=20)

# Set the x-axis to show every 90 minutes
plt.gca().xaxis.set_major_locator(mdates.MinuteLocator(interval=90))

# Set the date format to 'HH:MM'
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))

# Rotate x-axis labels and change fontsize
plt.xticks(rotation=25, fontsize=12)
plt.yticks(fontsize=12)

# Add gridlines 
plt.grid(True, which='both', axis='y', linestyle='--', linewidth=1.2, alpha=0.8)

# Set the background color of the grid area to white
plt.gca().set_facecolor('white')

# Customize the outside border to be a very light green
plt.gca().spines['top'].set_color('lightgreen')
plt.gca().spines['bottom'].set_color('lightgreen')
plt.gca().spines['left'].set_color('lightgreen')
plt.gca().spines['right'].set_color('lightgreen')

# Set the border width
plt.gca().spines['top'].set_linewidth(2)
plt.gca().spines['bottom'].set_linewidth(2)
plt.gca().spines['left'].set_linewidth(2)
plt.gca().spines['right'].set_linewidth(2)


# legend 
plt.legend(loc='upper left', fontsize=10, markerscale=1.5, bbox_to_anchor=(1.05, 1))

# Adjust layout 
plt.tight_layout()


# Show the plot
plt.show()





```
